```python
import pandas as pd
```

```python
df = pd.read_csv('./data/koreagram_insta_crawling.csv', encoding='utf-8')
df
```

```python
w = list(df['tags'])
for i in range(150):
    w[i] = w[i].strip("[""]")
w[0]
```

```python
s = ' '.join(w)
s
```

```python
from nltk.tokenize import RegexpTokenizer
# 세 문자 이상의 단어만 추출
tokenizer = RegexpTokenizer("[\w']{3,}")

# 소문자 변환 후 토큰화 실행
tokens = tokenizer.tokenize(s.lower())
tokens
```

```python
tokens_koreagram = []
for token in tokens:
    tokens_koreagram.append(token.strip("'"))
tokens_koreagram
```

```python
from nltk.corpus import stopwords
english_stops = set(stopwords.words('english'))
result_word = [word for word in tokens_koreagram if word not in english_stops]
result_word
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
from wordcloud import WordCloud
from collections import Counter

count = Counter(result_word)
tags = count.most_common(100)
tags
```

```python
wc = WordCloud(font_path = r'C:\Windows\Fonts\malgun.ttf',background_color='white').generate_from_frequencies(dict(tags))
plt.figure(figsize=(15,8))
plt.imshow(wc)
plt.axis('on')
plt.show()
```
