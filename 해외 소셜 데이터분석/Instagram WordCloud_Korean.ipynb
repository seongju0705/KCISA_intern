```python
import pandas as pd
```

```python
df = pd.read_csv('./data/koreagram_insta_crawling.csv', encoding='utf-8')
df
```

```python
w = list(df['tags'])
for i in range(150):
    w[i] = w[i].strip("[""]")
w[0]
```

```python
from konlpy.tag import Okt, Hannanum, Kkma, Komoran
okt = Okt()
hannanum = Hannanum()
kkma = Kkma()
komoran = Komoran()
```

```python
s = w[0]
print()

print("okt:",okt.nouns(s))
print("hannanum:",hannanum.nouns(s))
print("kkma:",kkma.nouns(s))
print("komoran:",komoran.nouns(s))
```

```python
up_total = ' '.join(w)
up_total
nouns_up = komoran.nouns(up_total)
```

```python
nouns_up
```

```python
stopwords = ['그램', '스타', '한복남', '곳', '수', '것','날', '번', '시', '만', '한곳', '은', '100', '딸', '한복', '대여']
nouns_total = []

for i in nouns_up:
    if i not in stopwords:
        nouns_total.append(i)
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
from wordcloud import WordCloud
from collections import Counter

up_count = Counter(nouns_total)
tags = up_count.most_common(100)

# down_count = Counter(nouns_down)
tags
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
from wordcloud import WordCloud
from collections import Counter

up_count = Counter(nouns_total)
tags = up_count.most_common(100)

wc = WordCloud(font_path = r'C:\Windows\Fonts\malgun.ttf',background_color='white').generate_from_frequencies(dict(tags))
plt.figure(figsize=(15,8))
plt.imshow(wc)
plt.axis('on')
plt.show()
```
