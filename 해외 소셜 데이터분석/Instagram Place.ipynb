# Instagram Place

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('./data/koreagram_2021_2022_place_clean.csv', encoding='cp949')
df
```

```python
df_y = df[df['대상여부'] == 'Y']
df_y
```

```python
df_y['place_final'] = np.where(pd.notnull(df_y['place_content']) == True, df_y['place_content'], df_y['place'])
df_y
```

```python
df_y['place_final'].isnull().sum()
```

```python
df_n = df[df['대상여부'] == 'N']
df_n
```

```python
df_result = pd.concat([df_n, df_y])
df_result = df_result[['content', 'place_final', 'place', 'place_content', '대상여부', 'date', 'like', 'tags', '네이버위치명', '경도', '위도', 'count']]
df_result
```

```python
df_result.to_csv('./data/koreagram_2021_2022_place_practice.csv', index=False, encoding='utf-8-sig')
```
