# Instagram Data Visualization
- 행정동 기준 시각화

## 라이브러리

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px

plt.rcParams['font.size'] = 12
plt.rcParams['figure.figsize'] = (15, 8)
```

## 서울

```python
seoul = pd.read_csv('./data/서울_구_행정동.csv', encoding='cp949')
seoul.head()
```

### 자치구 순위

```python
seoul_gu = pd.DataFrame(seoul['구'].value_counts())
seoul_gu.reset_index(level=0, inplace=True)
seoul_gu.columns = ['구', 'count']
seoul_gu
```

```python
sns.set(font='Malgun Gothic', style='whitegrid')
sns.barplot(data=seoul_gu, x='구', y='count')

plt.title('서울시 자치구별 언급량', fontsize=20)
plt.xlabel('자치구', fontsize=15)
plt.ylabel('Count', fontsize=15)
plt.xticks(fontsize=12, rotation=90)
plt.show()
```

### 행정동 순위

```python
seoul = seoul.sort_values(by=['count'], ascending=False)
seoul = seoul.reset_index(drop=True)
seoul
```

```python
seoul_top10 = seoul[0:10]
seoul_top10
```

```python
sns.set(font='Malgun Gothic', style='whitegrid')
sns.barplot(data=seoul_top10, x='동', y='count')

plt.title('서울시 행정동별 언급량 (Top 10)', fontsize=20)
plt.xlabel('행정동', fontsize=15)
plt.ylabel('Count', fontsize=15)
plt.xticks(fontsize=12, rotation=90)
plt.show()
```
