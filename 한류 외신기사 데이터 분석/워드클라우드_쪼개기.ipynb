```python
import pandas as pd
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt
import re
import konlpy
```

### 데이터 불러오기

```python
df = pd.read_csv('./전체.csv', encoding='cp949')
df
```

```python
df1 = df.loc[43501:45313]
df1
```

```python
w = list(df1['내용'])
w[0]
```

### 한국어

```python
from konlpy.tag import Komoran
komoran = Komoran(max_heap_size=1024*6)
```

```python
kor = []
for i in range(len(w)):
    kor.append(komoran.nouns(w[i]))
```

```python
import itertools
kor = list(itertools.chain.from_iterable(kor))
kor
```

```python
kor_stopwords = ['것', '수', '년', '말', '등']
kor_total = []

for i in kor:
    if i not in kor_stopwords:
        kor_total.append(i)
```

```python
from wordcloud import WordCloud
from collections import Counter

kor_count = Counter(kor_total)
kor_final = kor_count.most_common(500)

kor_final
```

### 영어

```python
text = ' '.join(w)
text
```

```python
eng_text = text.encode('utf-8').decode('ascii', 'ignore')
eng_text = re.sub('[^a-zA-Z ]', '', eng_text).strip()
eng_text
```

```python
from nltk.tokenize import RegexpTokenizer
tokenizer = RegexpTokenizer("[\w']{3,}")
eng = tokenizer.tokenize(eng_text)
eng
```

```python
eng_stopwords = ['the', 'that', 'and']
eng_total = []

for i in eng:
    if i not in eng_stopwords:
        eng_total.append(i)
```

```python
eng_count = Counter(eng_total)
eng_final = eng_count.most_common(500)

eng_final
```

### 한글 + 영어 합치기

```python
kor_df = pd.DataFrame(kor_final, columns = ['word', 'count'])
eng_df = pd.DataFrame(eng_final, columns = ['word', 'count'])

print(kor_df)
print(eng_df)
```

```python
final = pd.concat([kor_df, eng_df], axis=0)
final = final.sort_values('count', ascending=False)
final = final.reset_index(drop=True)
final
```

```python
final.to_csv('./45313.csv', encoding='cp949', index=False)
```
